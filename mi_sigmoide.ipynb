```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def sigmoide(t):   #Creacion de la funcion sigmoide.
    sigma=(1)/(1+np.exp(-t))
    return sigma
```

```python
x=np.linspace(-4,4)  #Creacion del vector de x
y=sigmoide(x)        #Obtener y en la funcion sigmoide.
```

```python
plt.scatter(0, 0.5, c='black')  #Graficar.
plt.plot(x, y)
plt.axhline(y=0.5, color='black', linestyle='--')
plt.grid(True)
plt.title('Funci√≥n Sigmoide')
plt.show()
```

```python
def sigmoide_general(t, w0, w1):   #Creacion de la funcion sigmoide general.
    sigma=1/(1+np.exp(-(w0+(w1*t))))
    return sigma
```

```python
x=np.linspace(0, 40)  #Agregar los parametros descritos en el documento
w0=[0, -4, -8, -12, -20]
w1=[0.4, 0.4, 0.6, 0.6, 1]
```

```python
plt.figure(figsize=(10, 6))  #Graficar cada evaluacion de los pesos al evaluarlos en la funcion sigmoide general.

for i, j in zip(w0, w1):
    y = sigmoide_general(x, i, j)
    plt.plot(x, y, label=f'w0={i}, w1={j}')

plt.title('Funci√≥n Sigmoide General')
plt.xlabel('X')
plt.ylabel('Y = s(x)')
plt.axhline(y=0.5, color='black', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

## Encontramos que los pesos ideales para la funcion sigmoide general estan entre los pesos [-12, 0.6] y [-20, 1]
