```python
cylinders=x[['vector','cylinders']].copy()    #Crear los dataframe con cada variable para su evaluacion.
displacement=x[['vector','displacement']].copy()
horsepower=x[['vector','horsepower']].copy()
weight=x[['vector','weight']].copy()
acceleration=x[['vector','acceleration']].copy()
model_year=x[['vector','model_year']].copy()
```

```python
cylinders
```

```python
cylinders=cylinders.to_numpy()  #Convertirlos a numpy array.
displacement=displacement.to_numpy()
horsepower=horsepower.to_numpy()
weight=weight.to_numpy()
acceleration=acceleration.to_numpy()
model_year=model_year.to_numpy()
```

```python
type(cylinders)
```

```python
def pesos(x,y):    #Funcion para calcular los pesos.
    w=np.linalg.inv(x.T@x)@x.T@y
    return w
```

```python
cyl_w=pesos(cylinders, y)            #Obetener los pesos de cada variable.
dis_w=pesos(displacement, y)
hor_w=pesos(horsepower, y)
wei_w=pesos(weight, y)
acc_w=pesos(acceleration, y)
mod_w=pesos(model_year, y)
```

```python
print('Pesos cylinders:', cyl_w)
print('Pesos displacement:', dis_w)
print('Pesos horsepower:', hor_w)
print('Pesos weight:', wei_w)
print('Pesos acceleration:', acc_w)
print('Pesos model_year:', mod_w)
```

```python
cyl_pred=cylinders@cyl_w  #Evaluar cada variable con su respectivo peso.
dis_pred=displacement@dis_w
hor_pred=horsepower@hor_w
wei_pred=weight@wei_w
acc_pred=acceleration@acc_w
mod_pred=model_year@mod_w
predicts=[cyl_pred,dis_pred,hor_pred,wei_pred,acc_pred,mod_pred]
```

```python
plt.scatter(x['cylinders'], y)
plt.scatter(x['cylinders'], cyl_pred)
plt.show()
```

```python
plt.scatter(x['model_year'], y)
plt.scatter(x['model_year'], mod_pred)
```

```python
fig, axes=plt.subplots(6,1,figsize=(10,40))    #Subplotear cada variable en una sola burbuja.
for i, col in enumerate(x.drop('vector', axis=1)):
    axes[i].scatter(x[col], y)
    axes[i].set_xlabel(col)
    axes[i].plot(x[col], predicts[i], '-r')
```

```python
ecm_cyl=((y-cyl_pred)**2).sum()/y.size   #Obtener el error cuadratico medio.
ecm_dis=((y-dis_pred)**2).sum()/y.size
ecm_hor=((y-hor_pred)**2).sum()/y.size
ecm_wei=((y-wei_pred)**2).sum()/y.size
ecm_acc=((y-acc_pred)**2).sum()/y.size
ecm_mod=((y-mod_pred)**2).sum()/y.size
```

```python
print('Error cuadratico medio cylinders:', ecm_cyl)
print('Error cuadratico medio displacement:', ecm_dis)
print('Error cuadratico medio horsepower:', ecm_hor)
print('Error cuadratico medio weight:', ecm_wei)
print('Error cuadratico medio acceleration:', ecm_acc)
print('Error cuadratico medio model_year:', ecm_mod)
```

## El modelo de ajuste del peso es el mas optimo, ya que es el que tiene un error cuadratico medio mas bajo, por lo que se puede decir que es mas preciso  (18.68).

```python
wei_df=x[['vector','weight']].copy()    #Crear los dataframes expandidos para la evaluacion.
wei_df["cuadrado"] = wei_df["weight"]**2
wei_cuad=wei_df.copy()
wei_df["cubico"]=wei_df["weight"]**3
wei_cub=wei_df.copy()
wei_df["cuartico"]=wei_df["weight"]**4
```
